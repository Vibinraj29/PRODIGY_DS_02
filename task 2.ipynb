```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic_data = pd.read_csv('Titanic.csv')
```

```python
titanic_data.head()
```

```python
titanic_data.isnull().sum()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(titanic_data.isnull(), yticklabels=False, cmap='viridis')
plt.title('Missing Values in Titanic Dataset')
plt.show()
```

```python
titanic_data['Age'].fillna(titanic_data['Age'].median(), inplace=True)

titanic_data['Cabin'].fillna(method='bfill', inplace=True)

titanic_data['Embarked'].fillna(method='bfill', inplace=True)
```

```python
titanic_data.isnull().sum()
```

```python
# Explore the relationship between survival and other variables (e.g., gender, class)
plt.figure(figsize=(10, 6))
sns.countplot(x='Survived', hue='Sex', data=titanic_data, palette='coolwarm')
plt.title('Survival Count by Gender')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Survived', hue='Pclass', data=titanic_data, palette='Set2')
plt.title('Survival Count by Passenger Class')
plt.show()
```

```python
# Investigate age distribution among passengers
plt.figure(figsize=(8, 6))
sns.histplot(data=titanic_data['Age'].dropna(), bins=30, kde=True, color='purple')
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Explore relationships between age, fare, and survival
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=titanic_data, palette='viridis')
plt.title('Age vs Fare with Survival')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```
